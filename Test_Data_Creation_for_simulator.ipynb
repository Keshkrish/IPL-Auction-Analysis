```python
import csv


with open('player list in order 2023.txt', 'r') as file:
    data = file.read().split('\n\n')


parsed_data = []
for line in data:
    # Remove numbers and split the string using '-'
    new_line=line.replace(' for ',' - ')
    new_line=new_line.replace('sold to ','')
    print(parts)
    parts = new_line.split(') ', 1)[1].split('-')
    if len(parts) == 3:
        player_name, team, price = parts
        price=price.lower()
        price = price.strip().replace('rs', '').replace('crore', '').strip()

        # Check if the price is in lakhs and divide by 100.0
        if 'lakh' in price:
            price = float(price.replace('lakh', '').strip()) / 100.0
            #print(line)
        else:
            price = float(price)

        parsed_data.append([player_name.strip(), team.strip(), price])
    elif len(parts)==2:
      parsed_data.append([parts[0].strip(),'Unsold',0.0])


csv_filename = 'output.csv'
with open(csv_filename, 'w', newline='', encoding='utf-8') as csv_file:
    csv_writer = csv.writer(csv_file)
    csv_writer.writerow(['Player Name', 'Team', 'Price'])
    csv_writer.writerows(parsed_data)

print(f'Data has been successfully parsed and saved to {csv_filename}')
```

```python
import copy
teams=['Chennai Super Kings', 'Delhi Capitals', 'Gujrat Titans', 'Kolkata Knight Riders', 'Lucknow Super Giants' ,'Mumbai Indians', 'Punjab Kings', 'Rajastan Royals' ,'Royal Challengers Banglore', 'Sunrisers Hyderabad']
fields=['purse','ba','bo','ar','wk','os']
x_train=pd.DataFrame()
for i in range(len(new_df)):
  purse_arr=[]
  role_arr=[]
  current=new_df.iloc[i]
  for t in teams:
    purse_arr.append(current[t+' purse'])
    role_arr.append(current[t+' '+current['role']])
  purse_arr=np.array(purse_arr)
  role_arr=np.array(role_arr)

  role_demand=role_max[current['role']]-role_arr

  for j in range(10):
    x_temp={}
    x_temp=copy.deepcopy(current)
    for t in teams:
      for f in fields:
        del x_temp[t+' '+f]
    x_temp['bid team']=teams[j]
    x_temp['no overseas']=current[teams[j]+' os']
    x_temp['demand_team']=role_demand[j]
    x_temp['supply']=current['supply']
    x_temp['demand_other']=np.sum(role_demand)-role_demand[j]
    x_temp['purse bid']=purse_arr[j]
    maxx_purse=0.0
    sum_purse=0.0
    for p in range(len(purse_arr)):
      if p!=j:
        sum_purse+=purse_arr[p]
        if maxx_purse<purse_arr[p]:
          maxx_purse=purse_arr[p]
    x_temp['max purse']=maxx_purse
    x_temp['avg purse']=(sum_purse-maxx_purse)/8.0

    if teams[j]==current['Team']:
      x_temp['output']=1
    else:
      x_temp['output']=0
    x_train=x_train.append(x_temp,ignore_index=True)

```
